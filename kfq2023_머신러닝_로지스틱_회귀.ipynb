```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
from sklearn.datasets import load_breast_cancer
```

```python
cancer = load_breast_cancer()
```

```python
data = cancer.data
label = cancer.target
```

```python
df=pd.DataFrame(data,columns = cancer.feature_names)
df
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
ss = StandardScaler()
scaled_data = ss.fit_transform(data)
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test = train_test_split(scaled_data,label)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
model = LogisticRegression()
model.fit(x_train,y_train)
pred = model.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score
from sklearn.metrics import roc_auc_score
from sklearn.metrics import recall_score
```

```python
print("acc:",accuracy_score(y_test,pred))
print("roc:",roc_auc_score(y_test,pred))
print("rec:",recall_score(y_test,pred))
```

```python
solovers = ['lbfgs',
            'liblinear',
            'newton-cg',
            'newton-cholesky',
            'sag',
            'saga']
for solver in solovers:
  model = LogisticRegression(solver=solver, max_iter=600)
  model.fit(x_train,y_train)
  pred = model.predict(x_test)
  print("====",solver,"====")
  print("acc:",accuracy_score(y_test,pred))
  print("roc:",roc_auc_score(y_test,pred))
  print("rec:",recall_score(y_test,pred))
```

```python
params = {
    'solver':['liblinear','lbfgs'],
    'penalty':['l2','l1'],
    'C':[0.01,0.1,1,5,10]
}
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
lr = LogisticRegression()
gs = GridSearchCV(lr,param_grid=params,scoring='accuracy',cv=3)
gs.fit(x_train,y_train)
print(gs.best_params_)
print(gs.best_score_)
```

```python
lr = LogisticRegression()
gs = GridSearchCV(lr,param_grid=params,scoring='recall',cv=3)
gs.fit(x_train,y_train)
print(gs.best_params_)
print(gs.best_score_)
```

```python
pred = gs.predict(x_test)
print("acc:",accuracy_score(y_test,pred))
print("roc:",roc_auc_score(y_test,pred))
print("rec:",recall_score(y_test,pred))
```

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import cross_val_score
```

```python
model = RandomForestRegressor(n_estimators=1000)
nmse = cross_val_score(model,x_train,y_train,
                scoring='neg_mean_squared_error',
                cv=5)
```

```python
print("neg mse:",nmse)
print("mse:",-1*nmse)
print("rmse:",np.sqrt(-1*nmse))
print("rmse 평균:",np.sqrt(-1*nmse).mean())
```

```python
path = '/content/drive/MyDrive/mldata/Boston_house.csv'
df = pd.read_csv(path)
data = df.drop('Target',axis=1)
label = df['Target']
ss = StandardScaler()
scaled_data = ss.fit_transform(data)
x_train,x_test,y_train,y_test = train_test_split(scaled_data,label)
```

```python
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import GradientBoostingRegressor
from xgboost import XGBRegressor
from lightgbm import LGBMRegressor
```

```python
m1 = DecisionTreeRegressor(max_depth=4)
m2 = RandomForestRegressor(n_estimators=1000)
m3 = GradientBoostingRegressor(n_estimators=1000)
m4 = XGBRegressor(n_estimators=1000)
m5 = LGBMRegressor(n_estimators=1000)
```

```python
models = [m1,m2,m3,m4,m5]
for model in models:
  print(model.__class__.__name__,"$$$$$$$$$$$$$$$$$$")
  nmse = cross_val_score(model,x_train,y_train,
                scoring='neg_mean_squared_error',
                cv=5)
  print("rmse 평균:",np.sqrt(-1*nmse).mean())
```
